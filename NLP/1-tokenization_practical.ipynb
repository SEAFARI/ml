```python
corpus = """Hello, welcome to NLP'S practical implementation. of tokenization.
Please follow the steps clearly to become expert in NLP!

"""
```

```python
print(corpus)
```

```python
## tokenization
## sentence --> paragraph
import nltk
from nltk.tokenize import sent_tokenize
nltk.download('all')
```

```python
document = sent_tokenize(corpus)
```

```python
type(document)
```

```python
for sentence in document:
    print(sentence)
```

```python
## TOKENIZATION
## paragraph --> words
## sentence --> words

from nltk.tokenize import word_tokenize
```

```python
word_tokenize(corpus)
```

```python
## for direct conversion

for sentence in document:
    print(word_tokenize(sentence))
```

```python
from nltk.tokenize import wordpunct_tokenize  ## splits punctuation also
```

```python
wordpunct_tokenize(corpus)
```

```python
from nltk.tokenize import TreebankWordDetokenizer  ## full stop is not be separated except the last word.
tokenizer  = TreebankWordDetokenizer()
```

```python
print(tokenizer.tokenize(corpus))
```
